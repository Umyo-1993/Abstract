```python

# Question



"""
ðŸ§ª Problem: Approximate a Solution Using Euler's Method

The differential equation is:
dy/dx = x + y
with initial condition y(0) = 1.

Task:
Approximate y at x=1 using Eulerâ€™s method with h=0.1.

No user input; just compute and print the result. Use Explicit Euler , Implicit Euler and RK 2
"""

def euler_method():
    x = 0       # initial x
    y = 1       # initial y
    h = 0.1     # step size
    x_end = 1   # end value of x

    while x < x_end:
        f = x + y        # f(x,y) = x + y
        y = y + h * f    # Euler update
        x = x + h

    print(f"Approximate value of y at x = {x_end} is: {y}")

euler_method()






#  Author U Mo
#  Dated 25/06/2025
#  Question


import matplotlib.pyplot as plt
x=0
y=1
h=0.1
xend=1
X=[]
Y=[]

while x < xend:
  f=x+y
  y=y+h*f  # Implementing Explicit Euler here
  Y.append(y) # storing the values in a list
  x=x+h
  X.append(x) # storing the values in a list

print("The values of X are " , X )
print ("The values of Y are ", Y)


 ## now ploting the values for X and Y ##
plt.plot(X,Y,marker='o')
plt.xlabel("values of X list")
plt.ylabel("values of Y list")
plt.title("Explicit Euler Method")
plt.show()




```

```python
 # solving using Implictit Euler
 ##################################################

import matplotlib.pyplot as plt

x = 0
y = 1
h = 0.1
xend = 1

X = []
Y = []

while x < xend:
    x_next = x + h
    y_next = (y + h * x_next) / (1 - h)  # Implicit Euler update

    X.append(x_next)  # Store the new x
    Y.append(y_next)  # Store the new y

    x = x_next       # Update x for next step
    y = y_next       # Update y for next step

print("The values of X are:", X)
print("The values of Y are:", Y)

# Plotting the values
plt.plot(X, Y, marker='o')
plt.xlabel("values of X list")
plt.ylabel("values of Y list")
plt.title("Implicit Euler Method")
plt.grid(True)
plt.show()



# change done for testing


```

```python
# way to solve using RK 02

import matplotlib.pyplot as plt
x = 0
y = 1
h = 0.1
xend = 1

X = [x]  # Store initial x
Y = [y]  # Store initial y

while x < xend:
    k1 = x + y
    k2 = (x + h/2) + (y + (h/2)*k1)
    y = y + h * k2
    x = x + h
    X.append(x)  # Store new x
    Y.append(y)  # Store new y

print("Approximate y at x=1:", y)

# Plotting (optional)
plt.plot(X, Y, marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('RK2 Approximation for dy/dx = x + y')
plt.grid(True)
plt.show()

```
