```python
# multistep methods
"""
Exercise: Solving an ODE using the 2-step Adams-Bashforth Method

Problem:
Given the ordinary differential equation (ODE):

    dy/dt = -2y + t^2

with initial condition:

    y(0) = 1

1. Use Euler’s method with step size h = 0.1 to compute y₁ (i.e., y at t = 0.1).
2. Then use the 2-step Adams-Bashforth method to compute y₂ (i.e., y at t = 0.2).

The 2-step Adams-Bashforth formula is:

    y_{n+1} = y_n + h/2 * [3*f(t_n, y_n) - f(t_{n-1}, y_{n-1})]

Tasks:
- Define the function f(t, y) = -2y + t²
- Compute y₁ using Euler’s method
- Compute y₂ using the 2-step Adams-Bashforth method
- Print the values of y₀, y₁, and y₂

Optional:
- Compare your result with the true value (if known)
"""

import matplotlib.pyplot as plt

# Step size
h = 0.1

# Initial condition
t0 = 0
y0 = 1

# Define the function f(t, y) = -2y + t^2
def f(t, y):
    return -2 * y + t**2

# Initialize t and y arrays
t = [t0]
y = [y0]

# Step 1: Use Euler’s method to compute y1
t1 = t0 + h
y1 = y0 + h * f(t0, y0)

t.append(t1)
y.append(y1)

print(f"y0 = {y[0]}")
print(f"y1 (Euler) = {y[1]}")

# Step 2: Use 2-step Adams-Bashforth to compute y2
t2 = t1 + h
y2 = y1 + h/2 * (3*f(t1, y1) - f(t0, y0))

t.append(t2)
y.append(y2)

print(f"y2 (Adams-Bashforth) = {y[2]}")

# Step 3: Compute 5 more steps using 2-step Adams-Bashforth
for i in range(3, 8):
    ti = t[i-1] + h
    yi = y[i-1] + h/2 * (3*f(t[i-1], y[i-1]) - f(t[i-2], y[i-2]))

    t.append(ti)
    y.append(yi)

# Print the values
for i in range(len(t)):
    print(f"t = {t[i]:.1f}, y = {y[i]:.6f}")

# Plot the results
plt.plot(t, y, marker='o', linestyle='-', color='blue')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Solution of dy/dt = -2y + t² using Adams-Bashforth Method')
plt.grid(True)
plt.show()







```

```python
# Local and Global Truncation Error
import math
import matplotlib.pyplot as plt # Import for plotting

# define the function at first
def function(t, y):
    """
    Represents the right-hand side of the ODE y'(t) = cos(t).
    """
    return math.cos(t)

# Define the exact solution function
def exact_solution(t):
    """
    Represents the exact solution to the ODE y'(t) = cos(t) with y(0)=0,
    which is y(t) = sin(t).
    """
    return math.sin(t)

# define the initial values and step size
t_0 = 0.0
y_0 = 0.0 # This is the initial numerical approximation, which is exact at t=0
h = 0.1

# --- Calculation for Local Truncation Error (LTE) - First Step ---
# This part is from our previous exercise, slightly adjusted for variable names
t_1 = t_0 + h
y_euler_prediction_from_exact_y0 = y_0 + h * function(t_0, y_0)
y_exact_at_t1 = exact_solution(t_1)
LTE = y_exact_at_t1 - y_euler_prediction_from_exact_y0

print("--- Local Truncation Error (LTE) for the First Step ---")
print(f"Time at next step (t_1): {t_1}")
print(f"Euler's prediction (from exact y_0): {y_euler_prediction_from_exact_y0}")
print(f"Exact solution at t_1: {y_exact_at_t1}")
print(f"Local Truncation Error (LTE): {LTE}\n")


# --- Calculation for Global Truncation Error (GTE) over time ---

# Define the total simulation time
t_end = 2.0 # Simulate up to t = 2.0

# Initialize lists to store results
time_points = []
numerical_solutions = []
exact_solutions = []
global_errors = []

# Set current time and solution for the loop
current_t = t_0
current_y_numerical = y_0

# Add initial values to lists
time_points.append(current_t)
numerical_solutions.append(current_y_numerical)
exact_solutions.append(exact_solution(current_t))
global_errors.append(abs(exact_solution(current_t) - current_y_numerical)) # Initial error should be 0

# Euler's Method Simulation Loop
while current_t < t_end:
    # Ensure we don't go past t_end if h causes an overshoot
    if current_t + h > t_end:
        h_actual = t_end - current_t # Adjust final step size
    else:
        h_actual = h

    # Calculate the next numerical solution using Euler's formula
    next_y_numerical = current_y_numerical + h_actual * function(current_t, current_y_numerical)

    # Move to the next time point
    current_t += h_actual
    current_y_numerical = next_y_numerical

    # Calculate the exact solution at the new time point
    y_exact_at_current_t = exact_solution(current_t)

    # Calculate the global error at this point
    current_global_error = abs(y_exact_at_current_t - current_y_numerical)

    # Append results to lists
    time_points.append(current_t)
    numerical_solutions.append(current_y_numerical)
    exact_solutions.append(y_exact_at_current_t)
    global_errors.append(current_global_error)

print("--- Global Truncation Error (GTE) Simulation Results ---")
# Optional: Print some of the results for verification
# for i in range(len(time_points)):
#     print(f"t={time_points[i]:.2f}, Num={numerical_solutions[i]:.6f}, Exact={exact_solutions[i]:.6f}, GTE={global_errors[i]:.6f}")

# Plotting the Global Error
plt.figure(figsize=(10, 6))
plt.plot(time_points, global_errors, marker='o', linestyle='-', color='red', label='Global Error')
plt.title('Global Truncation Error of Euler\'s Method over Time')
plt.xlabel('Time (t)')
plt.ylabel('Global Error (|Exact - Numerical|)')
plt.grid(True)
plt.legend()
plt.show()

# Optional: Plot numerical vs exact solution for visual comparison
plt.figure(figsize=(10, 6))
plt.plot(time_points, exact_solutions, label='Exact Solution', color='blue')
plt.plot(time_points, numerical_solutions, label='Numerical Solution (Euler)', linestyle='--', color='green', marker='x', markersize=4)
plt.title('Exact vs. Numerical Solution using Euler\'s Method')
plt.xlabel('Time (t)')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()
plt.show()
```
